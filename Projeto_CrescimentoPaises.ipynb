```python
import requests
import pandas as pd
```

# Brasil

```python

url_pib_brasil = "https://api.worldbank.org/v2/country/BRA/indicator/NY.GDP.MKTP.KD.ZG?date=2003:2011&format=json"


response = requests.get(url_pib_brasil)

data = response.json()

pib_brasil = pd.json_normalize(data[1])
```

```python
pib_brasil
```

```python
#Limpeza

pib_brasil = pib_brasil.drop(['unit', 'obs_status', 'countryiso3code', 'decimal', 'indicator.value', "country.id", "country.value", "indicator.id"], axis=1)

pib_brasil = pib_brasil.rename(columns={
    'date': 'Ano',
    'value': 'Cresc Pib em %'
})

pib_brasil = pib_brasil.replace('BRA', 'Brasil')

pib_brasil['Cresc Pib em %'] = pib_brasil['Cresc Pib em %'].round(2)

pib_brasil = pib_brasil.sort_values(by='Ano')

pib_brasil.set_index('Ano', inplace=True)
```

```python
pib_brasil
```

```python
url_ipca_brasil = "https://api.worldbank.org/v2/country/BRA/indicator/FP.CPI.TOTL.ZG?date=2003:2011&format=json"

response = requests.get(url_ipca_brasil)
data = response.json()

ipca_brasil = pd.json_normalize(data[1])
```

```python
#Limpeza

ipca_brasil = ipca_brasil.drop(['unit', 'obs_status', 'countryiso3code', 'decimal', 'indicator.value', "country.id", "country.value", "indicator.id"], axis=1)

ipca_brasil = ipca_brasil.rename(columns={
    'date': 'Ano',
    'value': 'Inflação em %'
})

ipca_brasil = ipca_brasil.replace('BRA', 'Brasil')

ipca_brasil['Inflação em %'] = ipca_brasil['Inflação em %'].round(2)

ipca_brasil = ipca_brasil.sort_values(by='Ano')

ipca_brasil.set_index('Ano', inplace=True)
```

```python
ipca_brasil
```

```python
url_desmp_brasil = "https://api.worldbank.org/v2/country/BRA/indicator/SL.UEM.TOTL.ZS?date=2003:2011&format=json"


response = requests.get(url_desmp_brasil)
data = response.json()

desmp_brasil = pd.json_normalize(data[1])
```

```python
#Limpeza

desmp_brasil = desmp_brasil.drop(['unit', 'obs_status', 'countryiso3code', 'decimal', 'indicator.value', "country.id", "country.value", "indicator.id"], axis=1)

desmp_brasil = desmp_brasil.rename(columns={
    'date': 'Ano',
    'value': 'Desemprego em %'
})

desmp_brasil = desmp_brasil.replace('BRA', 'Brasil')

desmp_brasil['Desemprego em %'] = desmp_brasil['Desemprego em %'].round(2)

desmp_brasil = desmp_brasil.sort_values(by='Ano')

desmp_brasil.set_index('Ano', inplace=True)
```

```python
desmp_brasil
```

```python
df_brasil = pib_brasil.join([ipca_brasil, desmp_brasil])

df_brasil.to_csv("tabela_brasil.csv", index = True, sep=",", encoding="utf-8")

df_brasil
```
